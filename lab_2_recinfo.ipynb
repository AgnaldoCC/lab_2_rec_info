```python

import pandas as pandas
import seaborn as seaborn
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import matplotlib.pyplot as plot_lib

nltk.download('stopwords')
stopwords = stopwords.words('portuguese')
```

```python
data_frame = pandas.read_csv('results.csv')

field_text = data_frame['text']

tokenizer = RegexpTokenizer(r'\b[^\d\W]+')

tokens = []

for document in field_text:
    tokens = tokens + tokenizer.tokenize(document.lower())

tokens_filter = []

for token in tokens:
    if (token not in stopwords and len(token) > 2):
        tokens_filter.append(token)

frequency = nltk.FreqDist(tokens_filter)
```

## 1. Reproduza a tabela 4.1 do livro texto calculando as estatísticas para sua coleção (5 pts).

```python
len_documents = len(field_text)
len_words = len(tokens_filter)
len_vocabulary = len(set(tokens_filter))
one_ocurrency_word = 0
more_1000_ocurrency_word = 0

for key in frequency.keys():
    if frequency[key] == 1000:
        more_1000_ocurrency_word += 1
    elif frequency[key] == 1:
        one_ocurrency_word += 1

table1 = pandas.DataFrame(data = {'Info': ['Total of Documents',
                                       'Total of Words',
                                       'Lenght of Vocabulary',
                                       'Words that occur one time',
                                       'Words that occur more than 1000 times'],
                              'Results': [len_documents,
                                          len_words,
                                          len_vocabulary,
                                          one_ocurrency_word,
                                          more_1000_ocurrency_word]})

table1
```

## 2. Reproduza a tabela 4.2 do livro considerando as top-50 palavras mais frequentes de sua coleção (5 pts).

```python
frequency_data_frame = pandas.DataFrame(frequency.most_common(50), columns = ['word', 'frequency'])
frequency_data_frame['r'] = frequency_data_frame['frequency'].rank(ascending = False)
frequency_data_frame['Pr'] = frequency_data_frame['frequency'] / len(tokens)
frequency_data_frame['Pr %'] = (frequency_data_frame['frequency'] / len(tokens) * 100)
frequency_data_frame['r.Pr'] = frequency_data_frame['r'] * (frequency_data_frame['frequency'] / len(tokens))
frequency_data_frame
```

## 3. Plote curvas de rank-frequência (usando escala logarítmica) para palavras e bigramas na coleção que você coletou na atividade anterior. Quais os melhores valores do parâmetro c para cada curva (5 pts)?

```python

diagrams = []

diagrams = nltk.bigrams(tokens_filter)

frequency_diagrams = nltk.FreqDist(diagrams)

diagrams_df = pandas.DataFrame(frequency_diagrams.most_common(), columns=['word', 'frequency'])
diagrams_df['r'] = diagrams_df['frequency'].rank(ascending = False)
```

```python
f, ax = plot_lib.subplots(figsize=(15, 7))
ax.set(xscale='log', yscale='log')
seaborn.regplot('r', 'frequency', diagrams_df, ax=ax, ci=68, scatter_kws={'s': 150}, color = 'b')
```
